```python
import networkx as nx

# Create a simple graph
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 1)])

# Calculate some basic properties
print(nx.average_shortest_path_length(G))
print(nx.degree_centrality(G))
```

```python
print(nx.betweenness_centrality(G))
print(nx.closeness_centrality(G))
print(nx.eigenvector_centrality(G))
```

# Layout Algorithms
- spring_layout: Force-directed layout (default)
- circular_layout: Nodes in a circle
- random_layout: Random positions
- shell_layout: Concentric circles
- kamada_kawai_layout: Force-directed, often more visually pleasing

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a sample graph
G = nx.karate_club_graph()  # Famous social network dataset

# Basic plotting
plt.figure(figsize=(10, 8))
nx.draw(G)
plt.show()
```

```python

# More detailed plotting with labels and custom appearance
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)  # Layout algorithm
nx.draw(G, pos,
        node_color='lightblue',    # Color nodes
        node_size=500,             # Size of nodes
        with_labels=True,          # Show node labels
        font_size=10,              # Label size
        edge_color='gray',         # Edge color
        width=1)                   # Edge width
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a sample graph
G = nx.karate_club_graph()
centrality = nx.degree_centrality(G)
pos = nx.spring_layout(G)

# Create figure and axes with space for colorbar
fig, ax = plt.subplots(figsize=(10, 8))

# Create a normalized color mapping
node_colors = list(centrality.values())
nodes = nx.draw(G, pos,
                node_color=node_colors,
                node_size=800,
                with_labels=True,
                cmap=plt.cm.viridis,
                edge_color='gray',
                ax=ax)

# Add colorbar with the correct normalization
sm = plt.cm.ScalarMappable(cmap=plt.cm.viridis, 
                          norm=plt.Normalize(vmin=min(node_colors), 
                                          vmax=max(node_colors)))
sm.set_array([])
plt.colorbar(sm, ax=ax, label='Degree Centrality')

plt.title('Network with Degree Centrality')
plt.tight_layout()
plt.show()
```
